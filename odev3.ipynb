```python
## /* 1. Asal Sayı Kontrolü (Kolay) */ ##

def is_asal(number):
    if number == 0 or number == 1:
        return False

    for i in range(2, number):
        if number % i == 0:
            return False
    return True

sayi = int(input("Bir tam sayı girin: "))

if is_asal(sayi):
    print(f"{sayi} asal bir sayıdır.")
else:
    print(f"{sayi} asal bir sayı değildir.")
```

```python
## /* 2. Palindromik Kelimeleri Bulma (Kolay) */ ##

def find_palindromes(word_list):
    palindrom = []

    for kelime in word_list:
        if kelime.lower() == kelime.lower()[::-1]:
            palindrom.append(kelime)

    return palindrom

print(find_palindromes(["level", "python", "radar", "kayak", "hello"]))
print(find_palindromes(["Anna", "world", "Madam"]))
```

```python
## /* 3. Fibonacci Sayılarını Üreten Fonksiyon (Orta) */ ##

def fibonacci(n):
    if n == 0:
        return []
    elif n == 1:
        return [0]

    dizi = [0, 1]

    for i in range(2, n):
        yeni_sayi = dizi[-1] + dizi[-2]
        dizi.append(yeni_sayi)

    return dizi

print(fibonacci(0))  # []
print(fibonacci(1))  # [0]
print(fibonacci(5))  # [0, 1, 1, 2, 3]
print(fibonacci(8))  # [0, 1, 1, 2, 3, 5, 8, 13]
```

```python
## /* 4. İki Listenin Kesişimini Bulma (Orta) */ ##

def intersection(list1, list2):

    ortak_elemanlar = list(set(list1) & set(list2))
    return ortak_elemanlar

girdi1 = input("Birinci listeyi girin (örn: 7,9,20): ")
list1 = [int(x.strip()) for x in girdi1.split(",")]

girdi2 = input("İkinci listeyi girin (örn: 7,9,20): ")
list2 = [int(x.strip()) for x in girdi2.split(",")]

print("Kesişim:", intersection(list1, list2))
```

```python
## /* 5. Fonksiyon Sonuçlarını Önbelleğe Alan Dekoratör (Zor) */ ##

def cache_results(func):
    cache = {}  # önbellek için sözlük

    def wrapper(*args):
        if args in cache:  # eğer daha önce çağrıldıysa
            print("Önbellekten döndü!")  # görmek için ekledim
            return cache[args]
        else:  # yoksa hesapla
            sonuc = func(*args)
            cache[args] = sonuc  # sonucu kaydet
            return sonuc

    return wrapper

# dekoratörü kullanıyoruz
@cache_results
def slow_fibonacci(n):
    if n < 2:
        return n
    return slow_fibonacci(n-1) + slow_fibonacci(n-2)

# testler
print(slow_fibonacci(10))  # ilk hesaplama
print(slow_fibonacci(10))  # önbellekten döner
```

```python
## /* 6. Çoklu Argüman Alan Matematiksel İşlem Fonksiyonu (Orta) */ ##

def hesapla(islem, *sayilar):
    if islem == "toplam":
        return sum(sayilar)
    elif islem == "carpim":
        sonuc = 1
        for sayi in sayilar:
            sonuc *= sayi
        return sonuc
    elif islem == "ortalama":
        return sum(sayilar) / len(sayilar)
    elif islem == "max":
        return max(sayilar)
    elif islem == "min":
        return min(sayilar)
    else:
        return None

# kullanıcıdan işlem türünü al
islem = input("Yapmak istediğiniz işlemi girin (toplam, çarpım, ortalama, max, min): ").strip().lower()

girdi = input("Sayıları girin (örn: 7,9,20): ")
sayilar = [float(x.strip()) for x in girdi.split(",")]

sonuc = hesapla(islem, *sayilar)
if sonuc is not None:
    print("Sonuç:", sonuc)
else:
    print("Geçersiz işlem türü!")
```

```python
## /* 7. Metin Analiz Fonksiyonu (Zor) */ ##

import string
from collections import Counter

def analyze_text(text):
    if not text.strip():
        return {
            "char_count": 0,
            "word_count": 0,
            "sentence_count": 0,
            "most_common_word": None,
            "average_word_length": 0.0
        }

    char_count = len(text)

    sentence_count = sum(text.count(p) for p in ".!?")

    temiz_text = text.lower().translate(str.maketrans("", "", string.punctuation))
    kelimeler = temiz_text.split()

    word_count = len(kelimeler)

    if kelimeler:
        most_common_word = Counter(kelimeler).most_common(1)[0][0]
    else:
        most_common_word = None

    if kelimeler:
        average_word_length = sum(len(kelime) for kelime in kelimeler) / word_count
    else:
        average_word_length = 0.0

    return {
        "char_count": char_count,
        "word_count": word_count,
        "sentence_count": sentence_count,
        "most_common_word": most_common_word,
        "average_word_length": round(average_word_length, 2)
    }

metin = input("Analiz etmek istediğiniz metni girin: ")

sonuc = analyze_text(metin)
print(sonuc)
```

```python
## /* 8. Banka Hesabı Sınıfı (Kolay) */ ##

class BankaHesabi:
    def __init__(self, hesap_sahibi):
        self.hesap_sahibi = hesap_sahibi
        self.bakiye = 0.0

    def para_yatir(self, miktar):
        if miktar > 0:
            self.bakiye += miktar
            print(f"{miktar} TL yatırıldı. Güncel bakiye: {self.bakiye} TL")
        else:
            print("Lütfen pozitif bir miktar girin.")

    def para_cek(self, miktar):
        if miktar <= 0:
            print("Lütfen pozitif bir miktar girin.")
        elif miktar > self.bakiye:
            print("Yetersiz bakiye!")
        else:
            self.bakiye -= miktar
            print(f"{miktar} TL çekildi. Güncel bakiye: {self.bakiye} TL")

    def bakiye_goster(self):
        print(f"Mevcut bakiye: {self.bakiye} TL")


# Kullanıcıdan hesap sahibi adı al
ad = input("Hesap sahibinin adını girin: ")
hesap = BankaHesabi(ad)

while True:
    # Menü seçeneklerini ayrı ayrı yazdırıyoruz
    print("Yapmak istediğiniz işlemi seçin:")
    print("1 - Para yatır")
    print("2 - Para çek")
    print("3 - Bakiye göster")
    print("4 - Çıkış")

    secim = input("Seçiminiz (1/2/3/4): ").strip()

    if secim == "1":
        miktar = float(input("Yatırmak istediğiniz miktarı girin: "))
        hesap.para_yatir(miktar)
    elif secim == "2":
        miktar = float(input("Çekmek istediğiniz miktarı girin: "))
        hesap.para_cek(miktar)
    elif secim == "3":
        hesap.bakiye_goster()
    elif secim == "4":
        print("Çıkış yapılıyor..")
        break
    else:
        print("Geçersiz seçim!")
```

```python
## /* 9. Öğrenci Not Sistemi (Kolay) */ ##

class Ogrenci:
    def __init__(self, ad, soyad):
        self.ad = ad
        self.soyad = soyad
        self.notlar = []

    def not_ekle(self, not_degeri):
        if 0 <= not_degeri <= 100:
            self.notlar.append(not_degeri)
            print(f"{not_degeri} eklendi. Mevcut notlar: {self.notlar}")
        else:
            print("Not 0 ile 100 arasında olmalıdır.")

    def ortalama_hesapla(self):
        if not self.notlar:
            return 0.0
        return round(sum(self.notlar) / len(self.notlar), 2)

    def harf_notu(self):
        ort = self.ortalama_hesapla()
        if 90 <= ort <= 100:
            return "AA"
        elif 80 <= ort < 90:
            return "BA"
        elif 70 <= ort < 80:
            return "BB"
        elif 60 <= ort < 70:
            return "CB"
        elif 50 <= ort < 60:
            return "CC"
        else:
            return "FF"


ad = input("Öğrencinin adını girin: ")
soyad = input("Öğrencinin soyadını girin: ")

ogrenci = Ogrenci(ad, soyad)

# kullanıcıdan notları al
while True:
    girdi = input("Eklemek istediğiniz notu girin (çıkmak için: '.'): ").strip()
    if girdi.lower() == '.':
        break
    try:
        not_degeri = float(girdi)
        ogrenci.not_ekle(not_degeri)
    except ValueError:
        print("Lütfen geçerli bir sayı girin:")

print(f"\n{ogrenci.ad} {ogrenci.soyad} için:")
print("Notlar:", ogrenci.notlar)
print("Ortalama:", ogrenci.ortalama_hesapla())
print("Harf Notu:", ogrenci.harf_notu())
```

```python
## /* 10. Kütüphane Yönetim Sistemi (Orta) */ ##

class Kitap:
    def __init__(self, baslik, yazar, isbn):
        self.baslik = baslik
        self.yazar = yazar
        self.isbn = isbn

    def __str__(self):
        return f"'{self.baslik}' by {self.yazar} (ISBN: {self.isbn})"

class Kutuphane:
    def __init__(self):
        self.kitaplar = []

    def kitap_ekle(self, kitap):
        self.kitaplar.append(kitap)
        print(f"Kitap eklendi: {kitap}")

    def kitap_ara(self, baslik):
        for kitap in self.kitaplar:
            if kitap.baslik.lower() == baslik.lower():
                return kitap
        return None

    def kitap_odunc_ver(self, baslik):
        kitap = self.kitap_ara(baslik)
        if kitap:
            self.kitaplar.remove(kitap)
            print(f"'{baslik}' ödünç verildi.")
        else:
            print(f"'{baslik}' kütüphanede bulunamadı.")

    def mevcut_kitaplari_goster(self):
        if not self.kitaplar:
            print("Kütüphanede kitap yok.")
        else:
            print("Kütüphanedeki kitaplar:")
            for kitap in self.kitaplar:
                print(f" - {kitap}")


kutuphane = Kutuphane()

while True:
    print("Yapmak istediğiniz işlemi seçin:")
    print("1 - Kitap ekle")
    print("2 - Kitap ara")
    print("3 - Kitap ödünç ver")
    print("4 - Mevcut kitapları göster")
    print("5 - Çıkış")

    secim = input("Seçiminiz: (1/2/3/4/5): ").strip()

    if secim == "1":
        baslik = input("Kitabın başlığını girin: ")
        yazar = input("Yazarını girin: ")
        isbn = input("ISBN numarasını girin: ")
        kitap = Kitap(baslik, yazar, isbn)
        kutuphane.kitap_ekle(kitap)
    elif secim == "2":
        baslik = input("Aramak istediğiniz kitabın başlığını girin: ")
        kitap = kutuphane.kitap_ara(baslik)
        if kitap:
            print(f"Bulundu: {kitap}")
        else:
            print(f"'{baslik}' kütüphanede bulunamadı.")
    elif secim == "3":
        baslik = input("Ödünç vermek istediğiniz kitabın başlığını girin: ")
        kutuphane.kitap_odunc_ver(baslik)
    elif secim == "4":
        kutuphane.mevcut_kitaplari_goster()
    elif secim == "5":
        print("Çıkış yapılıyr..")
        break
```

```python
## /* 11. Araç Kiralama Sistemi ve Kalıtım (Orta) */ ##

class Arac:
    def __init__(self, marka, model, gunluk_ucret):
        self.marka = marka
        self.model = model
        self.gunluk_ucret = gunluk_ucret

    def bilgileri_goster(self):
        print(f"Marka: {self.marka}")
        print(f"Model: {self.model}")
        print(f"Günlük Ücret: {self.gunluk_ucret} TL")

class Otomobil(Arac):
    def __init__(self, marka, model, gunluk_ucret, kapi_sayisi):
        super().__init__(marka, model, gunluk_ucret)
        self.kapi_sayisi = kapi_sayisi

    def bilgileri_goster(self):
        super().bilgileri_goster()
        print(f"Kapı Sayısı: {self.kapi_sayisi}")

class Motorsiklet(Arac):
    def __init__(self, marka, model, gunluk_ucret, motor_hacmi):
        super().__init__(marka, model, gunluk_ucret)
        self.motor_hacmi = motor_hacmi

    def bilgileri_goster(self):
        super().bilgileri_goster()
        print(f"Motor Hacmi: {self.motor_hacmi} cc")

print("Araç Kiralama Sistemi")

while True:
    print("1 - Otomobil ekle")
    print("2 - Motorsiklet ekle")
    print("3 - Çıkış")
    secim = input("Seçiminiz: ").strip()

    if secim == "1":
        marka = input("Otomobil markasını girin: ")
        model = input("Otomobil modelini girin: ")
        gunluk_ucret = float(input("Günlük ücreti girin: "))
        kapi_sayisi = int(input("Kapı sayısını girin: "))
        oto = Otomobil(marka, model, gunluk_ucret, kapi_sayisi)
        print("Otomobil Bilgileri:")
        oto.bilgileri_goster()

    elif secim == "2":
        marka = input("Motorsiklet markasını girin: ")
        model = input("Motorsiklet modelini girin: ")
        gunluk_ucret = float(input("Günlük ücreti girin: "))
        motor_hacmi = int(input("Motor hacmini (cc) girin: "))
        motor = Motorsiklet(marka, model, gunluk_ucret, motor_hacmi)
        print("Motorsiklet Bilgileri:")
        motor.bilgileri_goster()

    elif secim == "3":
        print("Çıkış yapılıyor..")
        break
    else:
        print("Geçersiz seçim!")
```

```python
## /* 12: E-ticaret Sistemi ve Kalıtım (Zor) */ ##

class Urun:
    def __init__(self, urun_id, isim, fiyat):
        self.urun_id = urun_id
        self.isim = isim
        self.fiyat = fiyat

    def bilgileri_goster(self):
        print(f"Ürün ID: {self.urun_id}")
        print(f"İsim: {self.isim}")
        print(f"Fiyat: {self.fiyat} TL")

class ElektronikUrun(Urun):
    def __init__(self, urun_id, isim, fiyat, garanti_suresi):
        super().__init__(urun_id, isim, fiyat)
        self.garanti_suresi = garanti_suresi

    def bilgileri_goster(self):
        super().bilgileri_goster()
        print(f"Garanti Süresi: {self.garanti_suresi} yıl")

class GiyimUrunu(Urun):
    def __init__(self, urun_id, isim, fiyat, beden, renk):
        super().__init__(urun_id, isim, fiyat)
        self.beden = beden
        self.renk = renk

    def bilgileri_goster(self):
        super().bilgileri_goster()
        print(f"Beden: {self.beden}")
        print(f"Renk: {self.renk}")

urun_listesi = []

while True:
    print("1 - Elektronik Ürün Ekle")
    print("2 - Giyim Ürünü Ekle")
    print("3 - Mevcut Ürünleri Göster")
    print("4 - Çıkış")
    secim = input("Seçiminiz: ").strip()

    if secim == "1":
        urun_id = int(input("Ürün ID girin: "))
        isim = input("Ürün ismini girin: ")
        fiyat = float(input("Fiyatı girin: "))
        garanti = int(input("Garanti süresi (yıl) girin: "))
        elektronik = ElektronikUrun(urun_id, isim, fiyat, garanti)
        urun_listesi.append(elektronik)
        print("Elektronik Ürün Eklendi:")
        elektronik.bilgileri_goster()

    elif secim == "2":
        urun_id = int(input("Ürün ID girin: "))
        isim = input("Ürün ismini girin: ")
        fiyat = float(input("Fiyatı girin: "))
        beden = input("Beden girin: ")
        renk = input("Renk girin: ")
        giyim = GiyimUrunu(urun_id, isim, fiyat, beden, renk)
        urun_listesi.append(giyim)
        print("Giyim Ürünü Eklendi:")
        giyim.bilgileri_goster()

    elif secim == "3":
        if not urun_listesi:
            print("Henüz ürün eklenmedi.")
        else:
            print("Mevcut Ürünler:")
            for urun in urun_listesi:
                print("------------------")
                urun.bilgileri_goster()

    elif secim == "4":
        print("Çıkış yapılıyor..")
        break

    else:
        print("Geçersiz seçim!")
```

```python
## /* 13: Dikdörtgen Sınıfı ve @property (Orta) */ ##

class Dikdortgen:
    def __init__(self, genislik, yukseklik):
        self.genislik = genislik
        self.yukseklik = yukseklik

    @property
    def genislik(self):
        return self._genislik

    @genislik.setter
    def genislik(self, deger):
        if deger <= 0:
            raise ValueError("Pozitif sayı girin!")
        self._genislik = deger

    @property
    def yukseklik(self):
        return self._yukseklik

    @yukseklik.setter
    def yukseklik(self, deger):
        if deger <= 0:
            raise ValueError("Pozitif sayı girin!")
        self._yukseklik = deger

    @property
    def alan(self):
        return self.genislik * self.yukseklik

    @property
    def cevre(self):
        return 2 * (self.genislik + self.yukseklik)

while True:
    try:
        genislik = float(input("Dikdörtgenin genişliğini girin: "))
        yukseklik = float(input("Dikdörtgenin yüksekliğini girin: "))
        dikdortgen = Dikdortgen(genislik, yukseklik)
        break
    except ValueError as e:
        print("Hata:", e)
        print("Pozitif sayıgirin!")

print("Dikdörtgen Bilgileri:")
print(f"Alan: {dikdortgen.alan}")
print(f"Çevre: {dikdortgen.cevre}")
```

```python
## /* 14. Çalışan Yönetim Sistemi ve Çok Biçimlilik (Zor) */ ##

from abc import ABC, abstractmethod

# Temel Çalışan sınıfı (Soyut)
class Calisan(ABC):
    def __init__(self, ad, soyad, calisan_id):
        self.ad = ad
        self.soyad = soyad
        self.calisan_id = calisan_id

    @abstractmethod
    def maas_hesapla(self):
        pass

# Tam zamanlı çalışan
class TamZamanli(Calisan):
    def __init__(self, ad, soyad, calisan_id, aylik_maas):
        super().__init__(ad, soyad, calisan_id)
        self.aylik_maas = aylik_maas

    def maas_hesapla(self):
        return self.aylik_maas

# Yarı zamanlı çalışan
class YariZamanli(Calisan):
    def __init__(self, ad, soyad, calisan_id, saat_ucreti, calisilan_saat):
        super().__init__(ad, soyad, calisan_id)
        self.saat_ucreti = saat_ucreti
        self.calisilan_saat = calisilan_saat

    def maas_hesapla(self):
        return self.saat_ucreti * self.calisilan_saat

# Tüm çalışanların toplam maaşını hesaplayan fonksiyon
def toplam_maas_hesapla(calisanlar):
    toplam = sum(calisan.maas_hesapla() for calisan in calisanlar)
    return toplam

# Kullanıcıdan input alarak çalışan ekleme
calisanlar = []

while True:
    print("Çalışan ekleme menüsü:")
    print("1 - Tam Zamanlı")
    print("2 - Yarı Zamanlı")
    print("3 - Çıkış ve toplam maaşı göster")
    secim = input("Seçiminiz: ").strip()

    if secim == "1":
        ad = input("Ad: ")
        soyad = input("Soyad: ")
        calisan_id = int(input("Çalışan ID: "))
        aylik_maas = float(input("Aylık maaş: "))
        calisanlar.append(TamZamanli(ad, soyad, calisan_id, aylik_maas))
        print(f"{ad} {soyad} eklendi (Tam Zamanlı).")

    elif secim == "2":
        ad = input("Ad: ")
        soyad = input("Soyad: ")
        calisan_id = int(input("Çalışan ID: "))
        saat_ucreti = float(input("Saatlik ücret: "))
        calisilan_saat = float(input("Çalışılan saat: "))
        calisanlar.append(YariZamanli(ad, soyad, calisan_id, saat_ucreti, calisilan_saat))
        print(f"{ad} {soyad} eklendi (Yarı Zamanlı).")

    elif secim == "3":
        if not calisanlar:
            print("Henüz çalışan eklenmedi.")
        else:
            print("Çalışan Maaşları:")
            for calisan in calisanlar:
                print(f"{calisan.ad} {calisan.soyad}: {calisan.maas_hesapla()} TL")
            print("Toplam Maaş:", toplam_maas_hesapla(calisanlar))
        break

    else:
        print("Geçersiz seçim!")
```
